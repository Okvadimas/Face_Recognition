```python
import numpy as np
import cv2
import pickle
```

```python
labels = {"person_name": 1}
with open("Data/Pickle/face-labels_dict.pickle", 'rb') as f:
    og_labels = pickle.load(f)
    labels = {v:k for k,v in og_labels.items()}
```

```python
import tensorflow as tf

new_model = tf.keras.models.load_model("Face_recognition_trained.h5")
new_model.summary()
```

```python
cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier("Data/OpenCV-Python-Series-master/src/cascades/data/haarcascade_frontalface_default.xml")

labels = {}
with open ("Data/Pickle/Face-labels_dict_cascade.pickle", "rb") as f:
    label = pickle.load(f)
    labels = {v:k for k,v in label.items()}
    
print(labels)

while True:
    bool, frame = cap.read()
    gray  = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
        
    for (x,y,w,h) in faces:
        face_img=gray[y:y+w,x:x+w]
        resized=cv2.resize(face_img,(224,224))
#         normalized=resized/255.0
        reshaped=np.reshape(resized,(-1,224,224,1))
        
        # bisa menggunakan predict classes atau numpy argmax untuk memprediksi
        result = new_model.predict_classes(reshaped)[0]
        print(result)
#         result=new_model.predict(reshaped)
#         print(result)
        
#         label=np.argmax(result)
#         print(label)
      
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255, 255, 255),2)
        cv2.rectangle(frame,(x,y-40),(x+w,y),(255, 255, 255),-1)
#         cv2.putText(frame, labels[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(0,255,0),2)
        cv2.putText(frame, labels[result], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(0,255,0),2)


    # Display the resulting frame
    cv2.imshow('LIVE',frame)
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
from tensorflow.keras.preprocessing import image
import os

test_dir = os.path.join("Data/face_test")

test_names = os.listdir(test_dir)
print(test_names) 
```

```python
def grayscale(img):
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return img
def equalize(img):
    img =cv2.equalizeHist(img)
    return img
def preprocessing(img):
    img = grayscale(img)
    img = equalize(img)
    img = img/255.0
    return img
def getCalssName(classNo):
    if   classNo == 0: print(labels[0])
    elif classNo == 1: print(labels[1])
    elif classNo == 2: print(labels[2])
    elif classNo == 3: print(labels[3])
    elif classNo == 4: print(labels[4])
    elif classNo == 5: print(labels[5])
    elif classNo == 6: print(labels[6])
                                        
```

```python
import numpy as np

for i in test_names:
    path = test_dir + i
    img = image.load_img(os.path.join(test_dir, i), target_size=(224,224), grayscale=True)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0) 
    x = x/255.0
    
    images = np.vstack([x])
    classes = new_model.predict(images, batch_size=10)
    print(classes)
    class_index = new_model.predict_classes(images)
    print(class_index)

    probabilityValue =np.amax(classes)

    input = int(class_index[0])
    
    print(str(round(probabilityValue*100,2) )+"%", end=" ")
    print(f"Image: {i} =", end=" ")
    getCalssName(input)
```
